Research Question # : Is there an association between Level of Intelligence and Age of first whole cigarette smoked? 


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

Codebook 1: Screen Shot 2023-04-27 at 11.59.38 AM.png

```python
#Manage data for variable 1
```

```python
df['H4MH7'].value_counts()
```

```python
df['intelligence_level'] = df['H4MH7']
df['intelligence_level'].value_counts()
```

```python
df['intelligence_level'] = df['intelligence_level'].loc[ df['intelligence_level'] !=' ']
df['intelligence_level'] = df['intelligence_level'].loc[ df['intelligence_level'] !='96']
df['intelligence_level'] = df['intelligence_level'].loc[ df['intelligence_level'] !='98']
df['intelligence_level'].value_counts()
```

```python
df['intelligence_level'].replace(['3', '4' , '5' , '6' , '2' , '1'] , ['about average' , 'slightly above average' , 'moderately above average' , 'extremely above average' , 'slightly below average' , 'moderately below average'] , inplace = True)
df['intelligence_level'].value_counts()
```

```python
#Manage data for variable 4
```

```python
df['H4PE4'].value_counts()
```

```python
df['mood_swings'] = df['H4PE4']
df['mood_swings'].value_counts()
```

```python
df['mood_swings'] = df['mood_swings'].loc[ df['mood_swings'] !=' ']
df['mood_swings'] = df['mood_swings'].loc[ df['mood_swings'] !='6']
df['mood_swings'] = df['mood_swings'].loc[ df['mood_swings'] !='8']
df['mood_swings'].value_counts()
```

```python
df['mood_swings'].replace(['4', '2' , '5' , '1' , '3'] , ['disagree' , 'agree' , 'strongly disagree' , 'strongly agree' , 'neither agree nor disagree'] , inplace = True)
df['mood_swings'].value_counts()
```

```python
rp.summary_cat(df['mood_swings'])
```

```python
#Graphs and tables or statistics
```

```python
plt.xlabel("Amount of Agreement")
plt.ylabel("Percent")
plt.title("I have frequent mood swings?")
df['mood_swings'].value_counts(normalize = True).plot(kind ='barh', color = 'Green', width = 0.2, alpha = 0.5, grid = True)
```

```python
plt.title("I have frequent mood swings?")
df['mood_swings'].value_counts().plot(kind ='pie', autopct='%1.1f%%', hatch=['**O',  '.||.'])
```
